```python
sector_counts = products["Sector(s)"].value_counts()
labels = sector_counts[0:3].index
sizes  = sector_counts[0:3].values
```

```python

c1=c2=c3=0
for s in products["Sector(s)"]:
    if(not pd.isnull(s)):
        s = s.split(";")
        for i in range(len(s)):
            sub = s[i].strip()
            if(sub == 'PreK-12'): c1+=1
            if(sub == 'Higher Ed'): c2+=1
            if(sub == 'Corporate'): c3+=1

fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Sector Distribution', size = 20, font="Serif")
explode = (0.05, 0.05, 0.05)
labels = ['PreK-12','Higher Ed','Corporate']
sizes = [c1,c2, c3]
ax.pie(sizes, explode=explode,startangle=60, labels=labels,autopct='%1.2f%%', pctdistance=0.7, colors=["#3b528b", "#21918c", "#5ec962", "#440154" ])
ax.add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```

El sector en la educaión que mas productos tiene es el de PreK-12, correspondiente a mas de la mitad de estos, seguido de Higher Ed y por último corporate

```python
temp_sectors = products['Sector(s)'].str.get_dummies(sep="; ")
temp_sectors.columns = [f"sector_{re.sub(' ', '', c)}" for c in temp_sectors.columns]
products = products.join(temp_sectors)
products.drop("Sector(s)", axis=1, inplace=True)

del temp_sectors
```

```python
#Separar las funcion principal en main y sub
primary_split = products["Primary Essential Function"].str.split(" - ", n=1, expand=True)

products["primary_function_main"] = primary_split[0]
products["primary_function_sub"] = primary_split[1]
```

```python
products
```

```python
c1=c2=c3=0
for s in products["primary_function_main"]:
    if(not pd.isnull(s)):
        c1 += s.count("CM")
        c2 += s.count("LC")
        c3 += s.count("SDO")

fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Primary Function (Main)', size = 20, font="Serif")
explode = (0.05, 0.05, 0.05)
labels = ['CM','LC','SDO']
sizes = [c1, c2, c3]
ax.pie(sizes, explode=explode,startangle=60, labels=labels,autopct='%1.2f%%', pctdistance=0.7,colors=["#440154","#3b528b", "#21918c", "#5ec962" ])
ax.add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```

EL 75% de los productos tienen la funcion primaria de "Aprendizaje y Curriculum"
